### Ordered over the years (sales value)

```python
# Sales value over the years
b = df[['Year','Sales']]
ordered_pro_year = b.groupby('Year').sum()
ordered_pro_year.reset_index(inplace=True)
ordered_pro_year
```

```python
# Sales value over the years and regions
b = df[['Year','Sales', 'Region']]
ordered_pro_year = b.groupby(['Year','Region']).sum()
ordered_pro_year.reset_index(inplace=True)
ordered_pro_year.set_index(['Year','Region'],inplace=True)
ordered_pro_year.sort_values(by=['Year', 'Sales'], ascending=[True,False])
```

```python
# Sales value over the years and regions and states
b = df[['Year','Sales', 'Region', 'State']]
ordered_pro_year = b.groupby(['Year','Region','State']).sum()
ordered_pro_year.reset_index(inplace=True)
ordered_pro_year.set_index(['Year','Region'],inplace=True)
ordered_pro_year.sort_values(by=['Year', 'Region','Sales'], ascending=[True,True,False])
```

```python
# Top 20 - Sales value 
b = df[['Year', 'Sales', 'Region', 'State', 'City']]
ordered_pro_year = b.groupby(['Year','Region','State','City']).sum()
ordered_pro_year.reset_index(inplace=True)
ordered_pro_year.set_index(['Year'],inplace=True)
ordered_pro_year.sort_values(by=['Sales'], ascending=False).head(20)
```

### Ordered over the years (quantity)

```python
# Number of orders over the years
b = df[['Year','Order_ID']]
ordered_pro_year = b.groupby('Year').count()
ordered_pro_year.reset_index(inplace=True)
ordered_pro_year
```

```python
# Number of orders over the years and regions
b = df[['Year','Order_ID', 'Region']]
ordered_pro_year = b.groupby(['Year','Region']).count()
ordered_pro_year.reset_index(inplace=True)
ordered_pro_year.set_index(['Year','Region'],inplace=True)
ordered_pro_year.sort_values(by=['Year', 'Order_ID'], ascending=[True,False])
```

```python
# Number of orders over the years and regions and states
b = df[['Year','Order_ID', 'Region', 'State']]
ordered_pro_year = b.groupby(['Year','Region','State']).count()
ordered_pro_year.reset_index(inplace=True)
ordered_pro_year.set_index(['Year','Region'],inplace=True)
ordered_pro_year.sort_values(by=['Year', 'Region','Order_ID'], ascending=[True,True,False])
```

```python
# Top 20 - Number of orders 
b = df[['Year', 'Order_ID', 'Region', 'State', 'City']]
ordered_pro_year = b.groupby(['Year','Region','State','City']).count()
ordered_pro_year.reset_index(inplace=True)
ordered_pro_year.set_index(['Year'],inplace=True)
ordered_pro_year.sort_values(by=['Order_ID'], ascending=False).head(20)
```

#### What products are most often sold together?

```python
sold_together = df[df['Order_ID'].duplicated(keep=False)]
sold_together['Together'] = sold_together.groupby('Order_ID')['Product_Name'].transform(lambda x: ','.join(x))
sold_together = sold_together[['Order_ID','Together']].drop_duplicates()
sold_together.head(10)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in sold_together['Together']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

count.most_common(20)

for key, value in count.most_common(20):
    print(key, value)
```
